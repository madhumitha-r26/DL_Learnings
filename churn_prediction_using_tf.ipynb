```python
import tensorflow as tf
import matplotlib.pyplot as plt   #data visualization
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Flatten,Dense
```



*  flatten - flattens multidimensional array into 1D array that import into dense layers
* dense - creates fully connected to the dense layer for the neural network




```python
(train_images,train_labels),(test_images,test_labels)=tf.keras.datasets.mnist.load_data() #mnist - handwriting dataset
train_images=train_images/255.0
test_images=test_images/255.0

print(train_images.shape)
print(test_images.shape)
print(train_labels)
```

displaying first image

```python
plt.imshow(train_images[0],cmap='grey')
plt.show()
```

define neural network model

```python
my_model=tf.keras.models.Sequential()
my_model.add(tf.keras.layers.Flatten(input_shape=(28,28)))
my_model.add(tf.keras.layers.Dense(128,activation='relu')) #reduce non-linearlity
my_model.add(tf.keras.layers.Dense(10,activation='softmax')) #softmax - probability of each class

my_model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
#adam - efficient optimizer

my_model.fit(train_images,train_labels,epochs=3) #epochs - makes the model to learn number of times (iterations)

val_loss,val_acc=my_model.evaluate(test_images,test_labels)
#val_loss - value loss on the model
#val_acc - model's accuracy

print("test accuracy of my model:",val_acc)
```
